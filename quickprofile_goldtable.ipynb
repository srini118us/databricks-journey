```python
# Quick profile of the gold table
spark.sql("""
SELECT
    COUNT(DISTINCT company_code)         AS n_company_codes,
    COUNT(DISTINCT currency)             AS n_currencies,
    MIN(cashflow_month)                  AS earliest_month,
    MAX(cashflow_month)                  AS latest_month,
    COUNT(*)                             AS total_rows
FROM workspace.default.gold_monthly_cashflow
""").show()

# Company codes present
spark.sql("""
SELECT company_code, currency,
       COUNT(*) AS months,
       ROUND(SUM(net_cashflow),0) AS total_net
FROM workspace.default.gold_monthly_cashflow
GROUP BY company_code, currency
ORDER BY company_code
""").show(50)
```
