```python
# Stage 0c — describe the SAP tables Lab 1 will build on
targets = [
    ("bdc_share_cash_flow",        "cashflow",                          "cashflow"),
    ("bdc_share_cash_flow",        "cashflow",                          "cashflowforecast"),
    ("bdc_share_supplier",         "supplier",                          "supplier"),
    ("bdc_share_vendorperformance","s4_zvendorperformance_dp_srv:v1",    "s4custom_vendorperformance"),
]

for cat, sch, tbl in targets:
    print("=" * 70)
    print(f"TABLE: {cat}.{sch}.{tbl}")
    try:
        df = spark.sql(f"DESCRIBE TABLE `{cat}`.`{sch}`.`{tbl}`")
        for r in df.collect():
            print(f"   {r[0]:<45} {r[1]}")
        cnt = spark.sql(f"SELECT COUNT(*) FROM `{cat}`.`{sch}`.`{tbl}`").collect()[0][0]
        print(f"   --> row count: {cnt}")
    except Exception as e:
        print(f"   ERROR: {e}")
```
