```python
# Stage 0b — list schemas and tables inside each SAP bdc_share_* catalog
sap_catalogs = [
    "bdc_share_cash_flow", "bdc_share_costcenter", "bdc_share_customer",
    "bdc_share_glaccount", "bdc_share_journal_entry", "bdc_share_supplier",
    "bdc_share_vendorperformance", "cashflow_data_product", "company_code_data_product"
]

for cat in sap_catalogs:
    print("=" * 60)
    print("CATALOG:", cat)
    try:
        schemas = spark.sql(f"SHOW SCHEMAS IN `{cat}`").collect()
        for s in schemas:
            sch = s[0]
            try:
                tables = spark.sql(f"SHOW TABLES IN `{cat}`.`{sch}`").collect()
                for t in tables:
                    print(f"   {sch}.{t[1]}")
            except Exception as e:
                print(f"   {sch}  (could not list tables: {e})")
    except Exception as e:
        print(f"   could not list schemas: {e}")
```
