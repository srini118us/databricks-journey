```python
# Lab 1 / Phase 1 / Stage B Part 1 — metadata for Genie accuracy
spark.sql("""
COMMENT ON TABLE workspace.default.gold_monthly_cashflow IS
'Monthly cash flow aggregated by company code, sourced from SAP BDC cash flow data product.
One row per company code, calendar month, and currency. Use for cash flow trends,
inflow/outflow analysis, and liquidity questions. Period covered: 2015 to 2025.'
""")

comments = {
    "company_code":      "SAP company code identifier - the legal entity (e.g. 1010, 1710). 18 distinct company codes.",
    "cashflow_month":    "First day of the calendar month for this cash flow summary.",
    "currency":          "Company code currency (ISO code: EUR, USD, GBP, AUD, CHF).",
    "transaction_count": "Number of cash flow transactions aggregated into this month.",
    "net_cashflow":      "Net cash flow for the month in company code currency (total_inflow + total_outflow). Positive means net cash in.",
    "total_inflow":      "Sum of all positive cash flow amounts (cash received) for the month, in company code currency.",
    "total_outflow":     "Sum of all negative cash flow amounts (cash paid out) for the month, in company code currency. Stored as a negative number.",
}
for col, desc in comments.items():
    spark.sql(f"COMMENT ON COLUMN workspace.default.gold_monthly_cashflow.{col} IS '{desc}'")

# verify
for r in spark.sql("DESCRIBE TABLE workspace.default.gold_monthly_cashflow").collect():
    print(f"  {r[0]:<20} {r[1]:<12} {r[2]}")
```
