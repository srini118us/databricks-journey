```python
# Find a catalog/schema we can write to
print("=== Catalogs visible ===")
for c in spark.sql("SHOW CATALOGS").collect():
    print(" ", c[0])

print("\n=== Schemas in 'workspace' ===")
try:
    for s in spark.sql("SHOW SCHEMAS IN workspace").collect():
        print(" ", s[0])
except Exception as e:
    print("  ERR:", e)

print("\n=== Current user / default catalog ===")
print("user:    ", spark.sql("SELECT current_user()").collect()[0][0])
print("catalog: ", spark.sql("SELECT current_catalog()").collect()[0][0])
print("schema:  ", spark.sql("SELECT current_schema()").collect()[0][0])

# Can we at least create a TABLE in workspace.default ?
print("\n=== Test: create table in workspace.default ===")
try:
    spark.sql("CREATE TABLE IF NOT EXISTS workspace.default._lab1_write_test (x INT)")
    print("  OK - can create tables in workspace.default")
    spark.sql("DROP TABLE IF EXISTS workspace.default._lab1_write_test")
except Exception as e:
    print("  NO:", str(e)[:200])
```
